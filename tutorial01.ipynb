```julia
using LinearAlgebra, Plots; pyplot();
```

```julia
f(x) = 1.0/(1.0 + exp(-x));
```

```julia
x_vals = -8.0:0.01:8.0;
y_vals = [f(x) for x in x_vals];
```

```julia
plot(x_vals,y_vals,xlabel="x",ylabel="f(x)",legend=false,lw=2)
```

Example of a simple node with one input and one output applied with three different weight values. 

```julia
using LaTeXStrings;
```

```julia
w = [0.5,1.0,2.0];
results = [f.(w.*x_vals) for w in w];
```

```julia
plot(x_vals,results,xlabel="x",ylabel=L"h_w(x)",label=[L"w_1=0.5" L"w_2=1.0" L"w_3=2.0"],lw=2)
```

Illustration of the effect of the bias term.

```julia
w = 5.0;
b = [-8.0,0.0,8.0];
results = [f.(w.*x_vals .+ b) for b in b];
```

```julia
plot(x_vals,results,xlabel="x",ylabel=L"h_w(x)",label=[L"b=-8.0" L"b=0.0" L"b=8.0"],lw=2)
```
